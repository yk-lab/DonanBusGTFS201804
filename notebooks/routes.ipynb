```python
import numpy as np
import pandas as pd
import collections
from pprint import pprint
import re
```

```python
weekday_master_timetable = pd.read_excel('../rawdata/weekday.xls', sheet_name=None, header=None)
weekend_master_timetable = pd.read_excel('../rawdata/weekend.xls', sheet_name=None, header=None)
```

```python
routes = dict()
print('# 平日')
weekday_routes = dict()
weekday_none = list()
for _k in weekday_master_timetable.keys():
    if weekday_master_timetable[_k].shape[1] > 2:
        a = weekday_master_timetable[_k][0:2].values
        if re.match(r'^1', _k):
            print('市内線', _k, a[1][1], a[0][5], end='')
            if _k in ['107500', '107510']:
                routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '新線', 'sub': ''}
                weekday_routes[_k] = {'v': a[0][5], 'route': '新線', 'sub': ''}
            elif re.match(r'^15', _k):
                print('(花火)', end='')
                #routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '花火大会', 'sub': ''}
                weekday_routes[_k] = {'v': a[0][5], 'route': '花火大会', 'sub': ''}
            else:
                b = re.search(r'([\d]+)-([\d]+)', a[1][1])
                if b.group(1) == '1401':
                    print('(学)', end='')
                    #routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '学生便', 'sub': b.group(2)}
                    weekday_routes[_k] = {'v': a[0][5], 'route': '学生便', 'sub': b.group(2)}
                else:
                    routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': b.group(1), 'sub': b.group(2)}
                    weekday_routes[_k] = {'v': a[0][5], 'route': b.group(1), 'sub': b.group(2)}
            print()
    else:
        weekday_none.append(_k)
        weekday_routes[_k] = {}

print('# 土日祝')
weekend_routes = dict()
weekend_none = list()
for _k in weekend_master_timetable.keys():
    if weekend_master_timetable[_k].shape[1] > 2:
        a = weekend_master_timetable[_k][0:2].values
        if re.match(r'^1', _k):
            print('市内線', _k, a[1][1], a[0][5], end='')
            if _k in ['107500', '107510']:
                routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '新線', 'sub': ''}
                weekend_routes[_k] = {'v': a[0][5], 'route': '新線', 'sub': ''}
            elif re.match(r'^15', _k):
                print('(花火)', end='')
                #routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '花火大会', 'sub': ''}
                weekend_routes[_k] = {'v': a[0][5], 'route': '花火大会', 'sub': ''}
            else:
                b = re.search(r'([\d]+)-([\d]+)', a[1][1])
                if b.group(1) == '1401':
                    print('(学)', end='')
                    #routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': '学生便', 'sub': b.group(2)}
                    weekend_routes[_k] = {'v': a[0][5], 'route': '学生便', 'sub': b.group(2)}
                else:
                    routes[_k] = {'v': a[0][5].replace('-', '〜'), 'route': b.group(1), 'sub': b.group(2)}
                    weekend_routes[_k] = {'v': a[0][5], 'route': b.group(1), 'sub': b.group(2)}
            print()
    else:
        weekend_none.append(_k)
        weekend_routes[_k] = {}
```

```python
gtfs_routes = pd.read_csv('../gtfs_201704/routes.txt', index_col=0)
gtfs_routes
```

```python
for i in range(len(gtfs_routes)):
    m = re.search(r'([^:]*)\:?\s?(.+)', gtfs_routes.iloc[i, 2])
    print(m.group(1), m.group(2))
    gtfs_routes.iloc[i, 1] = m.group(1)
    gtfs_routes.iloc[i, 2] = m.group(2)
```

```python
for i in range(len(gtfs_routes)):
    gtfs_routes.iloc[i, 2] = re.sub(r'\((.*)〜室蘭港\)', r'(\1〜室蘭フェリーターミナル)', gtfs_routes.iloc[i, 2])
    gtfs_routes.iloc[i, 2] = re.sub(r'\(室蘭港〜(.*)\)', r'(室蘭フェリーターミナル〜\1)', gtfs_routes.iloc[i, 2])
    gtfs_routes.iloc[i, 2] = re.sub(r'室蘭港', r'フェリー', gtfs_routes.iloc[i, 2])
```
