```python
weekdayTripNum = dict()
for i in weekday_master_timetable.keys():
    j = weekday_master_timetable[i].iat[0, 1]
    if not isinstance(j, float) or not math.isnan(j):
        m = re.search(r'便数:([\d]+)', j)
        if m:
            weekdayTripNum[i] = int(m.group(1))
        else:
            print('Error:', j)
```

# 記号付きを取得

```python
weekday_master_timetable = pd.read_excel('../rawdata/weekday.xls', sheet_name=None, skiprows=2, index_col=1)
weekday_master_timetable = {i: weekday_master_timetable[i].dropna(how='all', axis=1) for i in weekday_master_timetable.keys()}
weekend_master_timetable = pd.read_excel('../rawdata/weekend.xls', sheet_name=None, skiprows=2, index_col=1)
weekend_master_timetable = {i: weekend_master_timetable[i].dropna(how='all', axis=1) for i in weekend_master_timetable.keys()}
```

```python
def service_id(i, sd, d, s, num):
    direction_id = None
    if i == 130410:
        direction_id = trips_2017.at[130310, 'direction_id']
    elif i == 107500:
        direction_id = trips_2017.at[107600, 'direction_id']
    elif i == 107510:
        direction_id = trips_2017.at[107610, 'direction_id']
    else:
        direction_id = trips_2017.at[i, 'direction_id']
    if isinstance(direction_id, np.ndarray):
        direction_id = direction_id[0]
    if s == '':
        return pd.DataFrame(
            [[sd, f'{d}_{i}_{n}', None, None, direction_id, None, None, 0, 0, None, None, None] for n in range(num)],
            columns=['service_id', 'trip_id', 'trip_headsign', 'trip_short_name', 'direction_id', 'block_id', 'shape_id', 'wheelchair_accessible', 'bikes_allowed', 'jp_trip_desc', 'jp_trip_desc_symbol', 'jp_office_id'],
            index=pd.Int64Index([i for n in range(num)], name='route_id')
        )
    elif s == '●':
        return pd.DataFrame(
            [[sd, f'{d}_{i}_school_{n}', None, None, direction_id, None, None, 0, 0, None, None, None] for n in range(num)],
            columns=['service_id', 'trip_id', 'trip_headsign', 'trip_short_name', 'direction_id', 'block_id', 'shape_id', 'wheelchair_accessible', 'bikes_allowed', 'jp_trip_desc', 'jp_trip_desc_symbol', 'jp_office_id'],
            index=pd.Int64Index([i for n in range(num)], name='route_id')
        )
```

```python
trips = pd.DataFrame()
sd = '20180401_weekday'
d = 'weekday'
for i in list(routes.index):
    s = list(map(lambda x: x.strip(), weekday_master_timetable[f'{i}'].loc['記号など'].values))
    if s != []:
#        print(i, end=' ')
#        print(s, end=' ')
        n = len(list(filter(lambda x: x == '', s)))
        ss = len(list(filter(lambda x: x == '●', s)))
#        print(n, end=', ')
#        print(ss, end=' ')
#        print(weekdayTripNum[f'{i}'] == n+ss)
#        print(service_id(i, sd, d, '', n))
#        print(service_id(i, sd, d, '●', ss))
        if n > 0:
            trips=trips.append(service_id(i, sd, d, '', n))
        if ss > 0:
            trips=trips.append(service_id(i, sd, d, '●', ss))
sd = '20180401_weekend'
d = 'weekend'
for i in list(routes.index):
    s = list(map(lambda x: x.strip(), weekend_master_timetable[f'{i}'].loc['記号など'].values))
    if s != []:
#        print(i, end=' ')
#        print(s, end=' ')
        n = len(list(filter(lambda x: x == '', s)))
        ss = len(list(filter(lambda x: x == '●', s)))
#        print(n, end=', ')
#        print(ss, end=' ')
#        print(weekendTripNum[f'{i}'] == n+ss)
#        print(service_id(i, sd, d, '', n))
#        print(service_id(i, sd, d, '●', ss))
        if n > 0:
            trips=trips.append(service_id(i, sd, d, '', n))
        if ss > 0:
            trips=trips.append(service_id(i, sd, d, '●', ss))
```
